```python
def student_grades(row):
    if row['Grade'] == "A":
        return "Excellent"
    elif row['Grade'] == "B":
        return "Good"
    elif row['Grade'] == "C":
        return "Average"
    elif row['Grade'] == "D":
        return "Poor"
    else:
        return "Needs improvement"
    
df["student_grade"] = df.apply(student_grades, axis=1)
df['student_grade']

df.head()
```

```python
#.apply()
def calculate_total(row):
    return row['Midterm_Score'] + row['Final_Score'] + row['Assignments_Avg'] + row['Quizzes_Avg'] + row['Participation_Score'] + row['Projects_Score']

df['Total Score'] = df.apply(calculate_total, axis=1)
df['Total Score']
```

```python
def calculate_average(row):
    return (row['Attendance (%)'] + row['Midterm_Score'] + row['Final_Score'] + row['Assignments_Avg'] + row['Quizzes_Avg'] + row['Participation_Score'] + row['Projects_Score']) / 7

df['Average_Score'] = df.apply(calculate_average, axis=1)
df['Average_Score']
```

```python
#how to categorize students based on their average score using .apply and .map
def categorize_students(row):
    if row['Average_Score'] >= 70:
        return "Excellent"
    elif row['Average_Score'] >= 55:
        return 'Good'
    elif row['Average_Score'] >= 40:
        return 'Average'
    else:
        return 'Needs Improvement'
    
df['Category'] = df.apply(categorize_students, axis=1)
df['Category']
df.head()
```

## Lambda Functions

- Anonymous functions that are defined using the lambda keyword.
- Small, one-time and inline functions.

Syntax of a lambda function:
lambda parameters:function body

```python
add_two = lambda x:x + 2
add_two(11)

sum_of_three_numbers = lambda x, y, z: x + y + z
result  = sum_of_three_numbers(1,10,12)
result

sum_of = lambda a: a + 2
result = sum_of(2)
result

max_of_3 = lambda x,y,z : max(x,y,z)
max_of_3(4,7,8)

concat = lambda a, b: a + " " + b
concat("Hello", "World!")
```

```python
##Combining Dataframes
#Using concatenation to combine dataframes

df1 = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df2 = pd.DataFrame({'A':[7,8,9],'B':[10,11,12]})
df3 = pd.DataFrame({'A':[13,14,15],'B':[16,17,18]})

#Concatenate along rows

result = pd.concat([df1,df2,df3], axis = 0)
result
```

```python
#Concatenate along columns

result = pd.concat([df1,df2,df3], axis = 1)
result
```

```python
#JOINS
#Inner joins - Returns only the records with matching keys in both tables
#Outer joins - Returns all records when theres is a match in at least one of the tables.
#Left joins  - Return all records from the left table and matching rows from the right table .. Unmatched rows will have nan
#Right joins - Return all records from the right table and matching rows from the left table .. Unmatched rows will have nan
```

```python
df1 = pd.DataFrame({'key': ['A','B','C'], 'value1':[1,2,3]})
df2 = pd.DataFrame({'key': ['B','C','D'], 'value1':[4,5,6]})

#inner join
inner_join = pd.merge(df1,df2, on='key', how='inner')
print('Inner Join:\n', inner_join)

#outer join
outer_join = pd.merge(df1, df2, on='key', how='outer')
print('Outer Join:\n', outer_join)

#left join
left_join = pd.merge(df1, df2, on='key', how='left')
print('Left Join:\n', left_join)

#right join
right_join = pd.merge(df1, df2, on='key', how='right')
print('Right Join:\n', right_join)
```
